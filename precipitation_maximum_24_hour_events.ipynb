```python
import pandas as pd
```

# First dataset

```python
raw_data = pd.read_csv('data/97016_1909_2024_prec.csv')

# Analyze the stations' data in order to choose one station for analysis
raw_data.groupby(['STATION', 'NAME']).agg({'DATE': ['min', 'max']}) 
```

```python
df = raw_data[raw_data['NAME']=='CLATSKANIE, OR US']
```

```python
# Look at the top 10 24-hour rainfall events
df.sort_values('PRCP', ascending=False)[:10]
```
